```python
#Carregando os dados
import pandas as pd
wine = pd.read_csv('wine_dataset.csv')
```

```python
#Renomeando o nome das colunas
wine.rename(columns={'fixed_acidity':'Ácido Fixo', 'volatile_acidity':'Ácido Volátil', 'citric_acid':'Ácido Crítrico', 'residual_sugar':'Açúcar Residual', 'chlorides':'Cloretos', 'free_sulfur_dioxide':'Dióxido de Enxofre Livre', 'total_sulfur_dioxide':'Dióxido de Enxofre Total', 'density':'Densidade', 'sulphates':'Sulfatos', 'alcohol':'Álcool', 'quality':'Qualidade', 'style':'Estilo'}, inplace = True)
```

```python
#Transformando alguns valores em números manipuláveis - Replace: Substituir
#Estilo dos vinhos: Vermelho será 0 e Branco será 1

wine['Estilo'] = wine['Estilo'].replace('red', 0)
wine['Estilo'] = wine['Estilo'].replace('white', 1)

wine.head(100)
```

```python
#Transformar a coluna em uma variável
y = wine['Estilo']
#Transformar em uma variável toda a tabela MENOS a coluna, especificada no eixo 1
x = wine.drop('Estilo', axis = 1)
```

```python
#Criando treino e teste com SK Learn - Algoritimo para estudar a probailidade do vinho ser tinto ou vinho branco

from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesClassifier

#treinar e testar em 30% em proporção
x_treino, x_teste, y_treino, y_teste = train_test_split (x, y, test_size = 0.3)
#analisa as combinações entre linhas e colunas, tanto da tabela com estilo quanto sem.

#Criando o modelo - utilizar os modelos de treino
#model_selection é usado para dividir nossos dados em conjuntos de treinamento e 
#teste onde variáveis de recursos são fornecidas como entrada no método

#ARVORE DE CLASSICAFICAÇÃO
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)

#Depois do treino, faz-se o teste um a um e imprime o teste de estado aleatório q é usado para reprodutibilidade dos dados.
resultado = modelo.score(x_teste, y_teste)
print('probabilidade', resultado)
```
